```python
import numpy as np
import pandas as pd
heroes = pd.read_csv("heroes_information.csv")
```

```python
###instalar desde anaconda navigator######
from pandasql import *
```

```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
query2 = " SELECT * FROM heroes limit 10"
```

```python
pysqldf(query2)
```

```python
query3 = " SELECT name FROM heroes"
pysqldf(query3)
#Primer nombre del superhéroe 
```

```python
query5 = " SELECT DISTINCT(Publisher)  FROM heroes"
pysqldf(query5)
#Diferentes casas publicadoras
```

```python
query6 = " SELECT COUNT(Height)  FROM heroes WHERE Height  > 200 "
pysqldf(query6)
#Heroes que miden mas de 200cm
```

```python
query7 = " SELECT COUNT(Race)  FROM heroes WHERE Race  = 'Human' AND Height > 200 "
pysqldf(query7)
#Heroes humanos que miden mas de 200 metros
```

```python
query6 = " SELECT COUNT(Weight)  FROM heroes WHERE Weight  > 100 AND Weight <200"
pysqldf(query6)
#HEroes que pesan mas de 100 libras y menos de 200
```

```python
query9 = " SELECT COUNT('Eye color')  FROM heroes WHERE 'Eye color'  = 'blue' OR 'Eye color' = 'red' "
pysqldf(query9)
#SuperHéroes con ojos color azul o rojo
```

```python
query10 = " SELECT COUNT(Race)  FROM heroes WHERE Race  = 'Human' OR Race = 'Mutant' GROUP BY 'Eye color' = 'green' "
pysqldf(query10)
```

```python
query11 = " SELECT name  FROM heroes ORDER BY Race DESC"
pysqldf(query11)
#El nombre del primer heroe si ordeno la tabla por raza de manera descendente
```

```python
query12 = " SELECT Count(Gender)  FROM heroes Group BY Gender = 'Male'"
pysqldf(query12)
#Heroes masculinos y femeninmos
```

```python
query13 = " SELECT Count(Publisher)  FROM heroes  HAVING COUNT(name) >15"
pysqldf(query13)
#Cuantas casas publicadoras tienen mas de 15 superheroes
```
